# 🟰 Operadores de Atribuição

## 🔀 Atribuição Múltipla

Com Python, você pode atribuir um único valor a várias variáveis ao mesmo tempo. Isso permite inicializar várias variáveis de uma vez, que você mesmo pode reatribuir posteriormente no programa ou por meio da entrada do usuário.

```python
nota1 = nota2 = nota3 = 0

pos1, pos2 = 20, 40
```

## 🔃 Troca de valores de variáveis

É possível usar o recurso de atribuição múltipla para fazer a troca de valores entre variáveis:

```python
primeiro = 1
ultimo = 10

primeiro, ultimo = ultimo, primeiro

print(primeiro, ultimo)
```

## ➕ Operadores de Incremento e Decremento

Os operadores de incremento e decremento são usados para adicionar ou subtrair valores de uma variável de forma simplificada. Eles são frequentemente utilizados para atualizar **contadores** ou realizar **iterações** em estruturas de controle.

- **Incremento**: Adiciona um valor à variável.
- **Decremento**: Subtrai um valor da variável.

Esses operadores ajudam a escrever código mais conciso e são representados por combinações específicas de símbolos em várias linguagens de programação.

## Operadores de Incremento e Decremento em Python

Python oferece operadores de atribuição composta, como `+=` e `-=`, para realizar as mesmas operações. Abaixo estão os operadores usados para incremento e decremento:

| Operador | Descrição                        | Exemplo                   |
|----------|----------------------------------|---------------------------|
| `+=`     | Incrementa e atribui o resultado | `x += 1` (equivale a `x = x + 1`) |
| `-=`     | Decrementa e atribui o resultado | `x -= 1` (equivale a `x = x - 1`) |

Exemplo:

```python
# Incremento em Python
x = 5
x += 1  # x agora é 6

# Decremento em Python
x -= 1  # x agora é 5 novamente
```


## Todos os Operadores de Atribuição composta

| Operador | Descrição                        | Exemplo                   |
|----------|----------------------------------|---------------------------|
| `+=`     | Incrementa e atribui o resultado | `x += 1` (equivale a `x = x + 1`) |
| `-=`     | Decrementa e atribui o resultado | `x -= 1` (equivale a `x = x - 1`) |
| `*=`     | Multiplica e atribui o resultado | `x *= 1` (equivale a `x = x * 1`) |
| `/=`     | Divide e atribui o resultado | `x /= 1` (equivale a `x = x / 1`) |
| `//=`     | Divisão inteira e atribui o resultado | `x //= 1` (equivale a `x = x // 1`) |
| `%=`     | Módulo e atribui o resultado | `x %= 1` (equivale a `x = x % 1`) |

Em Python, a manipulação de variáveis através de incrementos e decrementos é comum em diversas aplicações, como loops, contagens e cálculos iterativos.



> ⚠️ Diferentemente de outras linguagens como C, C# ou Java, Python **não possui** os operadores de incremento `++` ou decremento `--`.


# ✅ Operadores Lógicos

São usados para compor uma combinação lógica com vários testes lógicos. Por exemplo, caso você queira avaliar se o motorista **tem carteira** e se o veículo **está licenciado**:

```python
carteira = True
licenciado = True

print(f'O motorista está liberado? {carteira and licenciado}')
```

Ou caso a pessoa tenha que atender ao menos um dos requisitos como ser maior de idade ou estar acompanhada dos pais:

```python
idade = 25
acompanhada = False

print(f'A pessoa pode acessar? {idade >= 18 or acompanhada}')
```

## Operações em cadeia

Em Python possível comparar, de forma encadeada, vários testes lógicos. Exemplo:
```python
# verificar se um valor x está entre 10 e 20:
print(x >= 10 and x <= 20)

# alternativa:
print(10 <= x <= 20)
```

## 🆔 Operadores de Identidade

Semelhante aos operadores relacionais, são semanticamente mais legíveis e podem ser usados para comparar ou diferenciar, dois objetos:

- `is`
- `is not`

A diferença primordial do uso dos operadores de identidade para os operadores relacionais é que os primeiros avaliam a referência, ou seja, se eles são o mesmo objeto:

```python
print(10.0 == 10) # True
print(10.0 is 10) # False
```

Mais detalhes sobre a diferença entre o `==` e o `is` [aqui](https://towardsdatascience.com/whats-the-difference-between-is-and-in-python-dc26406c85ad).

> 💡 Na ordem de precedência, os operadores relacionais ficam a frente

## 🔀 `if` inline

Em Python, é possível controlar o fluxo de execução do código usando **estruturas condicionais**.

```python
if condicao:
  # se a 1ª condição for verdadeira
elif condicao2:
  # se a 2ª condição for verdadeira
elif condicao3:
  # se a 3ª condição for verdadeira
else:
  # caso nenhuma das anteriores seja verdadeira
```

Quando temos uma comparação mais simples, podemos usar o `if` _inline_, uma versão mais curta e legível do `if` tradicional.

```python
(caso verdadeiro) if (condição) else (caso falso)
```

> 💡 Pode ser usado para executar funções específicas ou para atribuir valores diferentes!

- **Index**: Cada caractere dentro de uma string pode ser acessado pela sua posição (índice). Esse *index* pode ser positivo ou negativo.

```python
py = 'Python'

```

- **Slicing**: Além de um caractere específico acessado pela posição, é possível acessar fazer o *fatiamento*. Fatiar permite que você obtenha uma **substring** da *string* original:

```python
txt[1:4] # caracteres da posição 1 (incluído) até 4 (excluído)
txt[2:5] # caracteres da posição 2 (incluído) até 5 (excluído)
txt[:5]  # caracteres da posição 0 (incluído) até 5 (excluído)
```

- **Slicing com passo**: Ainda existe uma outra notação que permite acessar *substrings* de uma posição até outra determinando o passo:

```python
txt[0:20:2] # caracteres da posição 0 (incluído) até 20 (excluído), a cada 2
```

> Obs.: *Não existe um tipo "caractere". Em Python um caracter é uma string que possui tamanho 1.*

## `f-strings`

`f-strings` (_formatted string literals_) surgiram no Python 3.6 e se tornaram o modo recomendado de interpolação de valores em strings.

- Sintaxe curta e legível;
- Avaliação inline de expressões;
- Recurso completo de formatação, equivalente ao método `str.format()` — porém mais rápido e menos verboso.

### Sintaxe básica

```python
curso = "NExT"
modulo = 2
print(f'Sou estudante do {curso} e estou no módulo {modulo}.')
```

### Expressões dentro de `f-strings`

Qualquer expressão válida de Python pode aparecer entre as chaves.

```python
preco = 19.90
quantidade = 4
print(f"Total = R$ {preco * quantidade}")
```

Também é possível acessar índices de listas:

```python
cores = ["vermelho", "verde", "azul"]
print(f"Cor escolhida: {cores[1]}")
```

### Especificadores de Formatação

Depois da expressão, use `:` para inserir um _mini-language_ de formatação.

```{ expressão : especificação }```

#### Alinhamento, Largura e Preenchimento

| Símbolo | Significado | Exemplo | Saída |
| :-----: | ----------- | ------- | ----- |
|   `<`   | Alinhar à esquerda  | `f'{"Python":_<10}'` | `Python____` |
|   `>`   | Alinhar à direita   | `f'{"Python":#>10}'` | `####Python` |
|   `^`   | Centralizar         | `f'{"Python":*^10}'`      | `**Python**`     |


#### Precisão de Casas Decimais

`{numero:.Nf}` – `N` casas decimais.

```python
pi = 3.1415926535
print(f"π ≈ {pi:.3f}")          # π ≈ 3.142
```

#### Separador de Milhar

Use `,` para inserir separadores conforme a `locale` inglesa.

```python
populacao = 12456789
print(f"População: {populacao:,}")   # 12,456,789
```

Para notação científica, use `e` ou `E`:

```python
print(f"{populacao:.2e}")            # 1.25e+07
```

#### Bases Numéricas

| Especificador | Resultado                   |
| ------------- | --------------------------- |
| `b`           | binário (`0b` omitido)      |
| `o`           | octal                       |
| `x` / `X`     | hexadecimal minúsc./maiúsc. |

```python
valor = 255
print(f"Bin: {valor:b}  Hex: {valor:X}")
```

### Textos Multilinha

É possível escrever um texto grande em várias linhas usando um `\` para intercalar os textos:

```python
txt = 'uma quantidade muito grande de texto'\
' que pode ser encadeada'\
' desta forma'
```

Para usar várias `f-strings`, é possível agrupá-las dentro de parênteses. O Python concatena literais adjacentes automaticamente:

```python
mensagem = (
    f"Nome : {nome}\n"
    f"Idade: {idade}"
)
print(mensagem)
```

### Depuração com o Operador `=`

Desde o Python 3.8, adicione `=` logo após a expressão para obter `“nome = valor”`.

```python
x = 42
print(f"{x=}")             # x=42
print(f"{x*2=}")           # x*2=84
```

### Uso de Percentual (`%`)

Na análise de dados, usamos porcentagem o tempo todo (taxas de conversão, churn, acurácia de modelos). O Python tem um especificador nativo para isso (`%`) que multiplica por 100 e adiciona o símbolo automaticamente:

```python
taxa_conversao = 0.05432
print(f"Taxa de conversão: {taxa_conversao:.2%}")
```

## Boas Práticas e Desempenho

> - Prefira `f-strings` a concatenação (`+` ou `%`) quando interpolar valores;
> - Use formatação explícita (`:.2f`, `:,`) para evitar surpresas em apresentação de dados;
> - Mantenha `f-strings` curtas. Se a expressão ficar longa, calcule o valor antes e apenas referencie a variável;
> - Evite lógica complexa (condicionais) dentro das chaves. Isso prejudica a legibilidade.

### Exercício 2

#### Atualização de Saldo Interativa (Atribuição Composta e `input`)

**Contexto:** Você está criando um pequeno script para simular as transações diárias de um usuário de aplicativo de pagamentos passo a passo, recebendo os valores digitados na tela.

**Tarefa:**

1. Peça para o usuário informar o saldo inicial da conta e armazene na variável `saldo`;
2. Pergunte qual foi o valor do PIX recebido no dia. Atualize a variável `saldo` somando esse valor usando o operador de incremento (`+=`);
3. Pergunte qual foi o valor do boleto pago no dia. Atualize a variável `saldo` subtraindo esse valor usando o operador de decremento (`-=`);
4. Imprima o saldo final formatado com o símbolo R$ e duas casas decimais, alinhado à direita ocupando 10 espaços.

````
saldo = float(input("Informe o saldo inicial: "))
pix = float(input("Informe o valor do PIX recebido: "))

saldo += pix

boleto = float(input("Informe o valor pago: "))
saldo -= boleto

print(f"R$ {saldo:>10.2f}")
````

### Exercício 4

#### Projeção de Crescimento (Exponenciação )

**Contexto:** O time de produto quer saber quantos usuários ativos a plataforma terá daqui a 3 meses, considerando uma taxa de crescimento constante sobre a base atual (juros compostos). A fórmula simplificada é:

$Projeção = Base \times (1 + Taxa)^3$

**Tarefa:**
1. Peça ao usuário a **base atual de usuários** (inteiro);
2. Peça ao usuário a **taxa de crescimento mensal** em formato decimal (ex: digite `0.05` para 5%);
3. Calcule a projeção usando o operador de potência `**`;
4. Use o `if` inline para alertar se a meta bater 10.000 usuários: armazene "Meta Atingida" se a projeção for maior ou igual a 10000, ou "Meta Não Atingida" caso contrário;
5. Imprima a projeção sem casas decimais e o status da meta.

````
base = int(input("Informe a base atual de usuários: "))
taxa = float(input("Informe a taxa de crescimento: "))

projecao = base * ((1 + taxa) ** 3)

status = "Meta atingida." if projecao >= 10000 else "Meta não atingida."

print(status)
print(f"{projecao:.0f}")
````

### Exercício 5

**Tarefa:**

1. Peça o código numérico do funcionário;
2. Peça a sigla do departamento (ex: MKT, DADOS, RH);

Exemplo de saída esperada: `ID: 00042 | Depto: MKT       .`

````
codigo_funcionario = int(input("Digite o código numérico do funcionário: "))
depto = input("Digite a sigla do departamento (ex: MKT, DADOS, RH): ")

resultado = f"ID: {codigo_funcionario:05d} | Depto: {depto:<10}"

print(resultado)
````
